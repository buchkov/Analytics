# В какой день после реги клиенты достигают чекпойнта ">10ттк"

В query.csv слепки данных клиентов на каждый день. 
Одна строка — один слепок данных. 
Статус клиента — текущий. 
Нужно посчитать кол-во клиентов на 1, 2, 3 … 20 день после регистрации, которые первый раз достигли чекпоинта в 10 и болееproducts_dish. 
Выводим клиентов в долях.


```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('query_result.csv')
data.head()
```

### Оставляем платных клиентов с products_dish > 9 и сортируем по id и дате

```python
data = data[data['products_dish'] > 9]
data = data[data['status'] == 'Платный']
data = data.sort_values(['client_id', 'date'], ascending = [True, True])
data.head()
```

### Удаляем дубликаты. Оставляем только первое действие клиента

```python
data.drop_duplicates('client_id', keep = 'first', inplace = True)
data.sort_values(['products_dish'], ascending = False)
```

### Удаляем клиентов, которым делали импорт/копирование меню

```python
data = data[ data['products_dish'] < 47 ]
data.sort_values(['products_dish'], ascending = False)
```

## Добавляем столбец с кол-вом дней после регистрации

```python
def day_number(row):
    """Подсчет количества дней с момента регистрации"""
    
    date_dt = datetime.strptime(row['date'], '%Y-%m-%d')
    
    date_reg_dt = datetime.strptime(row['date_reg'], '%Y-%m-%d %H:%M:%S')
    
    return (date_dt.date() - date_reg_dt.date()).days
```

Считаем количество дней между date_reg и date для каждой строки

```python
data['days_from_reg'] = data.apply(day_number, axis = 1)
data.head(10)
```

Считаем количество строк для каждого значения столбца days_from_reg

```python
data_grouped = data.groupby('days_from_reg').count().reset_index()
data_grouped
```

```python
data_grouped = data.groupby('days_from_reg').count().reset_index()
data_grouped
```

Оставляем дни после регистрации и кол-во

```python
data_grouped = data_grouped[ ['days_from_reg', 'client_id'] ]
data_grouped
```

Считаем сколько всего пользователей в таблице

```python
total_users = data_grouped['client_id'].sum()
total_users
```

Добавляем столбец с долями

```python
data_grouped.loc[:, 'percent'] = data_grouped['client_id'] / total_users
data_grouped
```
